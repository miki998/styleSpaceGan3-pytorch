### Visualization result after edits
The channels used here were found with settings (batch=1) due and found for specifically SG3R and SG3T architectures. We show below some visualization to have qualitative assessement.

```python
# loading edited image using LELSD compute directions (screenshots from streamlit)
hairO = cv2.imread('../images/goodEditHairO.png')
hairM = cv2.imread('../images/goodEditHairM.png')

noseO = cv2.imread('../images/goodEditNoseO.png')
noseM = cv2.imread('../images/goodEditNoseM.png')

eyeO  = cv2.imread('../images/deformEyeO.png')
eyeM  = cv2.imread('../images/deformEyeM.png')
```

Example of good edits obtained

```python
fig,ax = plt.subplots(1,4, figsize=(25,5))
ax[0].imshow(hairO[:,:,::-1])
ax[1].imshow(hairM[:,:,::-1])
ax[0].set_title('Good hair edit: Original')
ax[1].set_title('Good hair edit: Modified= -66')

ax[2].imshow(noseO[:,:,::-1])
ax[3].imshow(noseM[:,:,::-1])
ax[2].set_title('Good nose edit: Original')
ax[3].set_title('Good nose edit: Modified= 55')
```

Example of bad edits obtained

```python
fig,ax = plt.subplots(1,2, figsize=(13,5))
ax[0].imshow(eyeO[:,:,::-1])
ax[1].imshow(eyeM[:,:,::-1])
ax[0].set_title('Deformation on eye edit: Original')
ax[1].set_title('Deformation on eye edit: Modified= -66')
```
